```python
import numpy as np, scipy as sp
from scipy.special import gamma
from matplotlib import pyplot as plt
%matplotlib inline
```

 It looks like you need to explicitly import the gamma function from scipy.special above.

```python
def fact(x):
    if x <= 2:
        return x
    return x*fact(x-1)
```

```python
x = np.linspace(0,10,11)
plt.plot(x,[fact(i) for i in x])
```

```python
print(fact(3.1))
```

```python
x = np.linspace(0,10,20000)
```

```python
xfact = [fact(i) for i in x]
```

```python
plt.plot(x,xfact,label='our def')
```

```python
plt.plot(x,xfact,label='our def')
plt.plot(x,gamma(x+1),label='gamma')
plt.legend()
```

What happens if we zoom in around an integer, say 9.0?

```python
plt.plot(x,xfact,label='our def')
plt.plot(x,sp.special.gamma(x+1),label='gamma')
plt.legend()

plt.xlim((8.9,9.1))
plt.ylim((250000,500000))
```

And what does the difference between our function and the gamma function look like in general?

```python
plt.plot(x,xfact - gamma(x+1),label='diff')
plt.legend()

#plt.xlim((9.5,10))
```

What about on a log scale?

```python
x = np.linspace(1,10,20000)
xfact = [fact(i) for i in x]
plt.plot(x,np.log(xfact),label='our def')
plt.plot(x,np.log(gamma(x+1)),label='gamma')
plt.legend()
```
