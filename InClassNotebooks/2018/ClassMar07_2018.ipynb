Do something with gradients.

What's a gradient?


$\nabla \phi = \mathrm{grad} \phi = \vec{i}\frac{\partial \phi}{\partial x} +
\vec{j}\frac{\partial \phi}{\partial y} +
\vec{k}\frac{\partial \phi}{\partial z}$

Related to directional derivative.
Vector.
Function (i.e. takes in x,y,z).
Steepest path up hill.

```python
f = np.array([1, 2, 4, 7, 11, 16], dtype=float)
```

```python
plt.plot(f)
```

```python
np.gradient(f)
```

```python
plt.plot(np.gradient(f))
```
